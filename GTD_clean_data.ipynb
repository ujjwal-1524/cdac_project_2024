```python
#to check whether the data contain null or nan values 
print(new_df.isnull().sum())
```

```python
print(new_df["latitude"].isnull().sum())
print(new_df["longitude"].isnull().sum())
#to drop the row where longitude and latitude conatin null values
new_df = new_df[new_df['longitude'].notna()]

```

```python
# Fill NaN values in Geographic Information
new_df['Country'].fillna('Unknown', inplace=True)
new_df['Region'].fillna('Unknown', inplace=True)
new_df['provstate'].fillna('Unknown', inplace=True)
new_df['city'].fillna('Unknown', inplace=True)
new_df["location"].fillna("Unknown",inplace=True)
new_df["Nationality"].fillna("Unknown",inplace=True)

# Fill NaN values in Attack Characteristics
new_df['Attacktype'].fillna(new_df['Attacktype'].mode()[0], inplace=True)  # Use mode
#new_df['weaptype1_txt'].fillna(df['weaptype1_txt'].mode()[0], inplace=True)  # Use mode
#new_df['weapsubtype1'].fillna(new_df['weapsubtype1'].mode()[0], inplace=True)  # Use mode
new_df["weapdetail"].fillna("Unknown",inplace=True)
new_df["Weapsubtype"].fillna("Unknown",inplace=True)

# Fill NaN values in Target Information
new_df['Targtype'].fillna(new_df['Targtype'].mode()[0], inplace=True)  # Use mode
new_df['Target'].fillna("Unknown", inplace=True)  
# Fill NaN values in Perpetrator Information
new_df['Group_name'].fillna('Group_name', inplace=True)
new_df['motive'].fillna('Unknown', inplace=True)

# Fill NaN values in Impact Information
new_df['nkill'].fillna(0, inplace=True)  # Assuming missing values mean no kills

new_df['nkillter'].fillna(0, inplace=True)  # Assuming missing values mean no kills
new_df['nwound'].fillna(0, inplace=True)  # Assuming missing values mean no wounds
new_df['nwoundte'].fillna(0, inplace=True)

new_df['nwoundte'].fillna(0, inplace=True)  # Assuming missing values mean no wounds
new_df['property'].fillna('Unknown', inplace=True)
new_df['propextent'].fillna('Unknown', inplace=True)
new_df['propvalue'].fillna(0, inplace=True)  # Assuming missing values mean no reported value
new_df['propcomment'].fillna('Unknown', inplace=True)

# Fill NaN values in Additional Context
new_df['summary'].fillna('Unknown', inplace=True)
new_df['suicide'].fillna('Unknown', inplace=True)
new_df['multiple'].fillna('Unknown', inplace=True)
new_df["Criterion"].fillna(-1,inplace=True)
new_df['Alternative'].fillna('Unknown', inplace=True)
new_df['Corp'].fillna('Unknown', inplace=True)
new_df['Guncertain'].fillna(-9, inplace=True)
new_df['Claimmode'].fillna('Unknown', inplace=True)
new_df['nhostkid'].fillna('Unknown', inplace=True)
new_df['kidhijcountry'].fillna('Unknown', inplace=True)
new_df['nreleased'].fillna(0, inplace=True)
new_df['scite'].fillna('Unknown', inplace=True)
```
